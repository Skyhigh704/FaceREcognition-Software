```python
from simpletransformers.classification import ClassificationModel
```

```python
model2 = ClassificationModel('bert', 'sagorsarker/bangla-bert-base', num_labels=2, use_cuda=True, args={
    'reprocess_input_data': True,
    'use_cached_eval_features': False,
    'overwrite_output_dir': True,
    'num_train_epochs': 3,
    'silent': True
})
```

```python
model2.train_model(train)
```

```python
test = test.drop(columns='target')
test = test.drop(columns='type')
test.set_axis(['sentence','hate speech'],axis=1,inplace=True)
test.rename(columns={'sentence':'text','hate speech':'labels'},inplace = True)
```

```python
result, model_outputs, wrong_predictions = model2.eval_model(test)
bads = {}
for pred in wrong_predictions:
    if pred.label in bads:
        bads[pred.label] += 1
    else:
        bads[pred.label] = 1
print("wrong predictions:")
print(str(len(wrong_predictions)) + ' wrong out of ' + str(len(test)))
bads
```

```python
from bnlp import CleanText

clean_text = CleanText(
   fix_unicode=True,
   unicode_norm=True,
   unicode_norm_form="NFKC",
   remove_url=True,
   remove_email=True,
   remove_emoji=True,
   remove_number=True,
   remove_digits=False,
   remove_punct=False,
   replace_with_url="<URL>",
   replace_with_email="<EMAIL>",
   replace_with_number="<NUMBER>",
   replace_with_digit="<DIGIT>",
   replace_with_punct = "<PUNC>"
)
```

```python
df2=pd.read_csv('/content/drive/My Drive/Hate_speech_dataset/scrapped-u_tube_onlyComments.csv')
df2.drop(df2.iloc[:,1:17],inplace=True,axis=1)
df2.rename(columns={'snippet.topLevelComment.snippet.textDisplay':'text'},inplace=True)
df2['clean_text']=df2['text'].apply(clean_text)
df2.head()
```

```python
df2.dropna(subset=['clean_text'], inplace=True)
```

```python
df2.head(10)
```

```python

predictions = []

for index, row in df2.iterrows():
    # Extract the text data from the CSV file (assuming the column name is 'text')
    text_data = row['clean_text']
    if text_data and len(text_data) > 0:
        # Make predictions
        result, raw_outputs = model2.predict(text_data)

        # Append the predictions to the list
        predictions.append({"text": text_data, "prediction": result[0]})
    else:
        # Handle empty or very short cleaned text
        print(f"Skipping empty or very short text at index {index}")

# Convert the list of predictions to a DataFrame if needed
predictions_df2 = pd.DataFrame(predictions)

# Display or save the predictions as needed
print(predictions_df2)
```

```python
predictions_df2.to_csv('/content/drive/My Drive/Hate_speech_dataset/preprocessed_utube.csv', index=False)
```
